```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\Vemul\Downloads\Python Advance\27th -11-24--  Pandas Dataframe\27th -11-24--  Pandas Dataframe\DataFrame_ Pandas\data.csv")
df.head()
```

```python
#----- Relational Plots (2 types) -------
# 1. Scatter Plot
# 2. Line Plot
```

```python
# 1. Scatter Plot
# Plots individual data points on a 2D plane.
```

```python
sns.scatterplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 2. Line Plot
# Connects data points with lines.
sns.lineplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# --------- Categorical Plots (7 types) --------
# 1. Box Plot
# 2. Violin Plot
# 3. Strip Plot
# 4. Swarm Plot
# 5. Bar Plot
# 6. Count Plot
# 7. Point Plot
```

```python
# 1. Box Plot
# Displays distribution with quartiles and potential outliers.
sns.boxenplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 2. Violin Plot
# Combines box plot and density plot for categorical data.
sns.violinplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 3. Strip Plot 
# Displays individual data points as dots.
sns.stripplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 4. Swarm Plot
# Similar to a strip plot but avoids overlapping points.
sns.swarmplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 5. Bar Plot
# Shows the mean of a continuous variable for each category.
sns.barplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# 4. Count Plot
# Displays the count of occurrences of each category.
sns.countplot(x=df['BirthRate'], hue = 'IncomeGroup', data=df)
```

```python
# 7. Point Plot 
# Similar to line plot but with points for categorical data.
sns.pointplot(x=df['BirthRate'], y=df['InternetUsers'], hue = 'IncomeGroup', data=df)
```

```python
# ------- Distribution Plots (4 types) ------
# 1. Histogram
# 2. Kernel Density Estimate Plot (KDE)
# 3. ECDF Plot (Empirical Cumulative Distribution Function)
# 4. Rug Plot
```

```python
# 1. Histogram
# Displays the frequency distribution of a continuous variable by binning the data.
# Useful for visualizing the shape and spread of the data.

sns.histplot(df['InternetUsers'], bins=20, kde=False)
```

```python
sns.histplot(df['InternetUsers'], bins=20, kde=True)
```

```python
# 2. Kernel Density Estimate Plot
# A smoothed curve that represents the probability density function of the data.
# Great for understanding the distribution shape without binning

sns.histplot(df['InternetUsers'])
```

```python
# 3. ECDF Plot (Empirical Cumulative Distribution Function)
# Displays the cumulative distribution, showing the proportion of data points below each value.
# Useful for understanding the cumulative behavior of the data.

sns.ecdfplot(df['InternetUsers'])
```

```python
# 4. Rug Plot
# Displays individual data points along the x-axis, providing granular insights into the data's distribution.
# Often used with other plots for additional context.

sns.rugplot(df['InternetUsers'])
```

```python
# ----------# Matrix Plots (2 types) -----------
# 1. Heatmap 
# 2. Clustermap
df.head()
```

```python
# 1. Heatmap 
# Visualizes a data matrix with values represented as colors.
# Commonly used for correlation matrices, but can be used for any matrix-style data.
# Customizable color maps, annotations, and cell formatting

table = df.pivot_table(index= 'CountryName', columns='BirthRate', values='InternetUsers')
sns.heatmap(table, annot=True, cmap='YlGnBu', linewidths='0.5')
```

```python
# 2. Clustermap 
# Similar to a heatmap but includes hierarchical clustering of rows and columns.
# Helps to identify groups or clusters in the data based on similarity.
# Includes dendrograms for visualizing the structure of the data.

data = {
    'Name': ['Jessica Burton', 'Keith Hines', 'Milan Waller', 'Cristian Farrell', 'Cesar Potter'],
    'Salary': [800000, 500000, 300000, 53000, 89000],
    'Bonus': [200000, 80000, 240000, 400000, 120000],
    'Performance_Score': [95, 88, 90, 72, 80]
}

df = pd.DataFrame(data)

# Setting 'Name' as index for clustering
df_clust = df.set_index('Name')

# Creating a clustermap
sns.clustermap(
    df_clust, 
    cmap='coolwarm',  # Color scheme
    annot=True,       # Display values on the heatmap
    linewidths=0.5,   # Add lines between cells
    figsize=(8, 6)    # Size of the plot
)

plt.title('Clustermap of Employee Metrics', y=1.05)
plt.show()
```

```python
# -------------Regression Plots (2 types)-------
# 1.regplot
    # Simpler, quick way to plot a regression line with scatter points.
# 2.lmplot
    # More advanced, supports faceting (subplots for different categories), and flexible in handling categorical data.
```

```python
# 1.regplot
# Simpler, quick way to plot a regression line with scatter points.

sns.regplot(x=df['BirthRate'], y=df['InternetUsers'], scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}, data=df)
```

```python
# 2.lmplot
# More advanced, supports faceting (subplots for different categories), and flexible in handling categorical data.

data = {
    'Name': ['Jessica Burton', 'Keith Hines', 'Milan Waller', 'Cristian Farrell', 'Cesar Potter'],
    'Salary': [800000, 500000, 300000, 53000, 89000],
    'Bonus': [200000, 80000, 240000, 400000, 120000],
    'EID': [120000, 75000, 45000, 7950, 13350]
}

df = pd.DataFrame(data)

# lmplot
sns.lmplot(data=df, x='Salary', y='Bonus', height=6, aspect=1.5, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Linear Regression Plot: Salary vs Bonus')
plt.xlabel('Salary')
plt.ylabel('Bonus')
plt.show()
```

```python
# ----------Facet Plots (3 types)--------
# 1. FacetGrid
# 2. catplot
# 3. pairplot
```

```python
# 2. catplot
# Quick and easy categorical plots with options for box plots, bar plots, etc.
data = {
    'Name': ['Jessica Burton', 'Keith Hines', 'Milan Waller', 'Cristian Farrell', 'Cesar Potter'],
    'Salary': [800000, 500000, 300000, 53000, 89000],
    'Bonus': [200000, 80000, 240000, 400000, 120000],
    'EID': [120000, 75000, 45000, 7950, 13350],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Male']
}

df = pd.DataFrame(data)
sns.catplot(data=df, x='Gender', y='Salary', kind='bar', height=5, aspect=1)
plt.title('Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
```

```python
# 3. pairplot
# used to visualize pairwise relationships in a dataset, creating a grid of scatterplots for numerical columns. 
#It is particularly helpful to observe patterns, trends, or outliers across all numeric features.

data = {
    'Name': ['Jessica Burton', 'Keith Hines', 'Milan Waller', 'Cristian Farrell', 'Cesar Potter'],
    'Salary': [800000, 500000, 300000, 53000, 89000],
    'Bonus': [200000, 80000, 240000, 400000, 120000],
    'EID': [120000, 75000, 45000, 7950, 13350],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Male']
}

df = pd.DataFrame(data)

# Pairplot
sns.pairplot(df, vars=['Salary', 'Bonus', 'EID'], hue='Gender', height=2.5)
plt.show()
```

```python
# --------Time Series Plots (1 type)--------
# Time series plots visualize data points collected or recorded at successive points in time.
# They are ideal for spotting trends, patterns, and seasonal variations over a given period.

data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [500, 550, 600, 450, 580, 610, 560, 630, 620, 700],
    'Profit': [50, 65, 70, 40, 60, 75, 55, 80, 70, 90],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(data)

# Line plot for Sales over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Sales', marker='o', label='Sales')
sns.lineplot(data=df, x='Date', y='Profit', marker='s', label='Profit')
plt.title('Time Series Plot of Sales and Profit')
plt.ylabel('Amount')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
